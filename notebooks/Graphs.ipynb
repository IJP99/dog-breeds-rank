```python
df_lif = pd.read_csv("lifetime_cost.csv")
df_pop = pd.read_csv("popularity.csv")
```

```python
df = pd.merge(df_lif, df_pop, on='Breed')
```

```python
df["lifetime_cost"] = df["lifetime_cost"].str.strip()
```

```python
df["lifetime_cost"] = df["lifetime_cost"].str.replace(",","")
```

```python
df["lifetime_cost"] = df["lifetime_cost"].astype(int)
```

```python
df
```

```python
plt.figure(figsize=(10,4))
ax = sns.barplot(df, x = "lifetime_cost", y = "Breed",
           palette = "viridis")
ax.bar_label(ax.containers[0], fontsize=8);
```

```python
df["ranking"] = 100 - df["popularity_ranking"]
sns.barplot(df, x = "ranking", y = "Breed",
           palette = "viridis")
```

```python
df
```

```python
df = pd.merge(df, df_lon, on='Breed')
```
