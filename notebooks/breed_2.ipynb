```python
import pandas as pd
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df_dog = pd.read_csv("dogs_longer_lifespan.csv")
df_dog
```

```python
data = {
    'Breed': ['Border Terrier', 'Cairn Terrier', 'Siberian Husky', 'Welsh Springer Spaniel', 'English Cocker Spaniel'],
    'Lifespan': [15, 14, 12, 13, 10]  # Example lifespans
}
df_dog = pd.DataFrame(data)
plt.figure(figsize=(8, 6))
sns.barplot(x='Lifespan', y='Breed', data=df_dog, palette='viridis')
plt.xlabel('Lifespan (years)')
plt.ylabel('Breed')
plt.title('Lifespan of Different Dog Breeds')
plt.xticks(rotation=45)

# Save the plot as an image file
plt.savefig('dog_lifespan.png', bbox_inches='tight', pad_inches=1)

# Show the plot
plt.show()
```

```python
df_dog = pd.read_csv("dogs_most_common_healthp.csv")
df_dog
```

```python
data = {'ailment': ['hip problems', 'ear infection', 'skin allergy', 'dental disease', 'dry eye'], 
        'count': [25, 20, 15, 10, 5]}
df_dog = pd.DataFrame(data)

# Plot
plt.figure(figsize=(8, 6))
sns.barplot(x='ailment', y='count', data=df_dog)
plt.xlabel('Ailment')
plt.ylabel('Count')
plt.title('Most Common Health Problems')
plt.xticks(rotation=45)
plt.savefig('most_common_health_problems.png', bbox_inches='tight', pad_inches=1)
plt.tight_layout()
plt.show()
```
